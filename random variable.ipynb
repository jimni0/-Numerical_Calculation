```python
#나쁜 랜덤 = 일련의 패턴이 보이는 랜덤이다.
def bad_ran(x):
    seed = (259*x + 17)%65536
    ran = float(seed)/65536
    return ran, seed

n = 5000
seed = 567

random1x = np.zeros(n)
random1y = np.zeros(n)

for i in range(n):
    result = bad_ran(seed)
    random1x[i] = result[0]
    seed = result[1]

    result = bad_ran(seed)
    random1y[i] = result[0]
    
# good random

random2 = np.random.random([5000,5000])
random2x = random2[:,0]
random2y = random2[:,1]

plt.figure(figsize=(12,6))
plt.subplot(121)
plt.scatter(random1x, random1y, s=0.2)
plt.title('Bad random')
plt.figure(figsize=(12,6))
plt.subplot(121)
plt.scatter(random2x, random2y, s=0.2)
plt.title('good random')
```

```python
#정규 분포 (normal distribution)

mean, sigma = 0, 0.1
data = np.random.normal(mean, sigma, 1000)
print(data)
```

```python
plt.hist(data, 30'개씩', color='skyblue',density = 1) # density = 1 노멀라이즈하는 것
```

```python
count, bins, ignored = plt.hist(data, 30, color='skyblue',density=1) # density = 1 노멀라이즈하는 것
```

```python
print(bins)
```

```python
def gaussian(x, mean, sigma):
    coeff = 1./(sigma*np.sqrt(2.*np.pi))
    value = coeff * np.exp(-(x-mean)**2.0/(2.0*sigma**2))
    return value

count, bins, ignored = plt.hist(data, 30, color='skyblue',density=1) 

y = gaussian(bins, mean, sigma)

plt.plot(bins, y, color='r')
```

```python
def f(x):
    return 3./2.*x**3 + 11./8.*x**2 + 1./6.*x + 1./12.

#probability distribution function (PDF)
n = 20
x = np.linspace(0,1,n)
y = f(x)

plt.plot(x,y, '-', c='black', label='f(x):PDF')
plt.legend()
```

```python
from scipy.integrate import quad

def f(x):
    return 3./2.*x**3 + 11./8.*x**2 + 1./6.*x + 1./12.

#probability distribution function (PDF)
n = 20
x = np.linspace(0,1,n)
y = f(x)

#cummulative distribution function (CDF)
ycum = np.zeros(n)
for i in range(n):
    value = quad(f,0.0, x[i])[0]
    ycum[i] = value

#normalization -> [0,1] 맥시멈값으로 나눠주겟다
M = np.max(ycum)
ycum = ycum/M
    
X = []
n = 10000

for i in range(n):
    ynow = np.random.uniform(0,1)
    xnow = np.interp(ynow, ycum, x) #인터폴레이션 ycum, x 이 둘의 함수를 기준으로 인터폴레이션에서 와이나우에서 엑스나우 구하기 ㅅㅂ 뭐라노
    X.append(xnow)

plt.plot(x,y, '-', c='black', label='f(x):PDF')
plt.hist(X, bins=20, density=1, label='Sempled', color='skyblue')
plt.legend()
```
