```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
def my_interp(p0,p1,x):
    #x0=p0[0], y0=p0[1]
    #x1=p1[0], y1=p1[1]
    y = (p1[1]-p0[1])/(p1[0]-p0[0])*(x-p0[0])+p0[1]
    return y

p0=[1,10]
p1=[8,2]
result = my_interp(p0, p1, x=3.5)
print(result)
```

```python
x = np.linspace(0,2*np.pi,10)
y = np.sin(x)

plt.plot(x,y,'o',color='black')
plt.show()
```

```python
#위의 것을 interpolation 하고 싶다!

x = np.linspace(0,2*np.pi,10)
y = np.sin(x)

xnew = np.linspace(0,2*np.pi,50)
ynew = np.interp(xnew,x,y)
#linear interplation 해서 그럼

plt.plot(x,y,'o',color='black')
plt.plot(xnew,ynew,'-',color='red')
plt.show()
```

```python
#위의 것을 interpolation 하고 싶다!

x = np.linspace(0,2*np.pi,10)
y = np.sin(x)

xnew = np.linspace(0,2*np.pi,100)
ynew = np.interp(xnew,x,y)
#linear interpolation 해서 그럼

plt.plot(x,y,'o',color='black')
plt.plot(xnew,ynew,'o',color='red') 
plt.show()
```

```python
from scipy import interpolate
x = np.linspace(0,2*np.pi,10)
y = np.sin(x)
newfunc=interpolate.interp1d(x,y,kind='linear')

xnew = np.linspace(0,2*np.pi,50)
ynew = newfunc(xnew)
plt.plot(x,y,'o',color='black')
plt.plot(xnew,ynew,'-',color='red')
plt.show()
```

```python
from scipy import interpolate
x = np.linspace(0,2*np.pi,10)
y = np.sin(x)
newfunc=interpolate.interp1d(x,y,kind='cubic')

xnew = np.linspace(0,2*np.pi,50)
ynew = newfunc(xnew)
plt.plot(x,y,'o',color='black')
plt.plot(xnew,ynew,'-',color='red')
```

```python
x = np.linspace(0,2*np.pi,10)
y = np.sin(x)

plt.plot(x,y,marker='o',c='black')

xnew = np.linspace(0,2*np.pi,50)
for kind in ['linear','quadratic','cubic']:
    func = interpolate.interp1d(x,y,kind=kind)
    ynew = func(xnew)
    plt.plot(xnew,ynew,label=kind)

plt.legend(loc='upper right',fontsize=15)
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

def func(x,y):
    return x*np.exp(-x**2-y**2)

x = np.linspace(-2,2,20)
y = np.linspace(-2,2,20)

xgrid, ygrid =  np.meshgrid(x,y)
z = func(xgrid,ygrid)
#2-d projection
plt.imshow(z, cmap=plt.cm.viridis)
#3-d projection
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection = '3d')
ax.plot_wireframe(xgrid,ygrid,z,color='r')
```

```python
newfunc = interpolate.interp2d(x,y,z,kind='cubic')
xnew = np.linspace(-2,2,100)
ynew = np.linspace(-2,2,100)
znew = newfunc(xnew, ynew)

xnewgrid, ynewgrid =  np.meshgrid(xnew,ynew)

plt.imshow(znew, cmap=plt.cm.viridis)

fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection = '3d')
ax.plot_wireframe(xnewgrid,ynewgrid,znew,color='r')
```
